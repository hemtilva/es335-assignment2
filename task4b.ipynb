```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as T
import torchvision.io as io
import matplotlib.pyplot as plt
import random
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
img = io.read_image("task4.jpg").float() / 255.0
C, H, W = img.shape
```

```python
imp_np = img.permute(1,2,0).numpy()
plt.imshow(imp_np)
```

```python
patch1 = img[:, 0:50, 450:500]
patch1_show = patch1.permute(1,2,0).numpy()
plt.imshow(patch1_show) 
```

```python
ranks = [5,10,25,50]
epochs = 1000
lr = 1e-2
patch = patch1
```

```python
for r in ranks:
    recon = torch.zeros_like(patch)

    for colour in range(3):
        n,m = patch[colour].shape

        U = torch.randn(n,r,device = device, requires_grad= True)
        V = torch.randn(r,m, device = device, requires_grad= True)
        optimizer = optim.Adam([U,V], lr = lr)
        loss_function = nn.MSELoss()

        for epoch in range(epochs):
            optimizer.zero_grad()
            X_hat = U @ V
            loss = loss_function(X_hat, patch[colour].to(device))
            loss.backward()
            optimizer.step()

        recon[colour] = (U @ V).detach().cpu()
    recon = torch.clamp(recon,0,1)

    plt.imshow(recon.permute(1,2,0).cpu())
    plt.title(f"Reconstructed Patch r = {r}")
    plt.axis("off")
    plt.show()
```
